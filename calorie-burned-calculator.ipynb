```python
import numpy as np
import pandas as pd

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

import pickle
```

```python
# Reading the csv file
df = pd.read_csv(r"C:\Users\Sheewakoti-Manish\Documents\My-Codes\Calorie-Burned-Calculator\data.csv")
```

```python
# Independent Features
x = df[["Gender", "Age", "Height", "Weight", "Duration", "Heart_Rate", "Body_Temp"]]

# Dependent Features
y = df["Calories"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
# Creating out an Instance of a Model 
model = RandomForestRegressor()
```

```python
# Fitting out the Model 
model.fit(x_train, y_train)
```

```python
# Making out predictions 
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
# Checking out r2 score
r2_score(y_test, y_pred)
```

```python
# Serializing the Model using pickle 
pickle.dump(model, open("model.pkl", "wb"))
```
